```python
class MyClass:
  def __init__(self):
    self.public_variable = 10
    self._private_variable =20

  def public_method(self):
    print("This is public method")

  def __private_method(self):
    print("This is private method")

  def variable(self):
    return self._private_variable



obj = MyClass()

res = obj.variable()
print(res)

print(obj.public_variable)
obj.public_method()

print(obj.__private_variable)
obj.__private_method()
```

GET AND SET METHODS

```python
class MyClass:
    def __init__(self, private_variable):
        self.__private_variable = private_variable

    def get_private_variable(self):
        return self.__private_variable

    def set_private_variable(self, new_value):
        self.__private_variable = new_value

# Example usage
obj = MyClass(private_variable=42)

# Accessing the private variable using the getter
value = obj.get_private_variable()
print("Current value:", value)

# Modifying the private variable using the setter
obj.set_private_variable(100)

# Accessing the modified value using the getter
new_value = obj.get_private_variable()
print("Modified value:", new_value)

# Attempting to access the private variable directly
# This will raise an AttributeError
try:
    direct_access = obj.__private_variable
except AttributeError as e:
    print("AttributeError:", e)
```

NAME MANGALING

```python
class MyClass:
    def __init__(self):
        self.__private_variable = 42

    def __private_method(self):
        pass

# Outside the class, name mangling occurs
obj = MyClass()
print(obj._MyClass__private_variable)  # Accessing the mangled variable
obj._MyClass__private_method()
```

```python
class Employee:
  def __init__(self,name,salary):
    self.name =name
    self.__salary = salary

  def work(self):
    self.__salary += 300

  def show(self):
    print(self.name, self.__salary)

Emp1 = Employee("vasanth",2000)



Emp1.work()

Emp1.show()

Emp1.__salary = 400
print(Emp1.__salary)

```
